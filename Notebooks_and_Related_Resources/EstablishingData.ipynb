# How to Get the Data


Source Location: This dataset can be downloaded directly from the Billboard Melodic Music Dataset report (https://transactions.ismir.net/articles/168/files/66b30cacae2f2.pdf). 

Download Date: November 5, 2025

Steps to Reproduce: The file is on page 13 of the document linked above, under the "Supplementary File" bullet. Open the downloaded folder and navigate to the BiMMuDa folder within it. The csv file for each song is in the "bimmuda_metadata" folder inside BiMMuDa.

Source Material: This dataset includes the top five songs from each year between 1950 and 2022. These 371 songs obtained their rankings from the official Billboard Year-End Singles Charts.


Creation Process: The melodies for each song in the dataset (371 in total) were recorded manually and used to create 371 MIDI files for each song. These MIDI files were then broken down into less complex song portions to create more of these files. These song and melody files, in addition to scores and lyrics, are combined to make the full dataset.
 


Format: While majority of this dataset is in the form of MIDI files, the specific spreadsheet detailing each song is in the form of a csv file.

# COLS Table


```python
def create_cols_table(df):
    data_info = {
        'Column Name': df.columns,
        'Data Type': df.dtypes,
        'Non-Null Count': df.notnull().sum()
    }
    cols_table = pd.DataFrame(data_info)
    cols_table = cols_table.reset_index(drop=True)
    return cols_table

COLS = create_cols_table(SONGS)
display(COLS)
```
